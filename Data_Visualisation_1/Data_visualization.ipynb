```python
df = pd.read_csv("IRIS.csv")
```

```python
# 5 Create a new categorical column 'petal_size': small (<2), medium (2–5), large (>5).
import numpy as np
conditions = [ df["petal_length"] < 2 , (df["petal_length"] >=2) & (df["petal_length"] <=5 )  , df["petal_length"] >5      
]

choices = ["small", "medium", "large"]
df["petal_size"]=np.select(conditions, choices, default="unknow")
```

```python
df["petal_size"].value_counts()
```

```python
df.groupby("species")["petal_size"].value_counts()
```

```python
df.groupby("species").agg({
    "petal_length" : "mean",
    "petal_width" : "mean",
    "sepal_length" : "mean",
    "sepal_width" : "mean",
})
```

```python
df.groupby("species").agg({
    "petal_length" : "max",
    "petal_width" : "min",
    "sepal_length" : "mean",
    "sepal_width" : "std",
})
```
