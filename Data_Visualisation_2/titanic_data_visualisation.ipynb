```python
import numpy as np

conditions = [
    (df["age"] <= 12),
    (df["age"] > 12) & (df["age"] <= 19),
    (df["age"] > 19) & (df["age"] <= 59),
    (df["age"] >= 60)
]

choices = ["Child", "Teen", "Adult", "Senior"]

df["age_group"] = np.select(conditions, choices, default = "Unknown")
```

```python
by_age = df.groupby("age_group")["survived"].mean()*100
```

```python
print(by_age)
```

```python
survival_rate_class = df.groupby("class")["survived"].mean()*100
print(survival_rate_class)
```

```python
avg_fair = df.groupby("survived")["fare"].mean()
print(avg_fair)
```

```python
by_town = df.groupby("embark_town")["survived"].mean()*100
print(by_town)
```

```python
df["family_size"]= df["sibsp"]+df["parch"]+1

by_family_size= df.groupby("family_size")["survived"].mean()*100
print(by_family_size)
```
